```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Resmi i√ße aktar

```python
img=cv2.imread("sudoku.jpg",0)
img=np.float32(img)
print(img.shape)
plt.figure()
plt.imshow(img,cmap="gray")
plt.axis("off")
```

harris corner detection

bloksize kom≈üuluk boyutu , ksize kutu boyutu,k ise free parameter

```python
hcd=cv2.cornerHarris(img,blockSize=2,ksize=3,k=0.04)
plt.figure()
plt.imshow(hcd,cmap="gray")
plt.axis("off")
```

k√∂≈üelerin daha belirƒüin olmasƒ± i√ßin tespit edilen noktarƒ±n geni≈ületilmesi.

```python
hcd=cv2.dilate(hcd,None)
img[hcd>0.2*hcd.max()]=1
plt.figure()
plt.imshow(hcd,cmap="gray")
plt.axis("off")
```

shai tomassi detection

(img,ka√ß adet kenar tespit etmesi i√ßin,quality_level,minumu distande iki k√∂≈üe arasƒ±nda)

```python
img=cv2.imread("sudoku.jpg",0)
img=np.float32(img)
corners=cv2.goodFeaturesToTrack(img,100,0.01,10)
corners=np.int64(corners)
```

```python
for i in corners:
    #d√ºzle≈ütirmek i√ßin kullanƒ±lƒ±r.
    x,y=i.ravel()
    # circle √ßizdir.(resim,kordinat,yarƒ±cap,renk,i√ßinidoldur)
    cv2.circle(img,(x,y),3,(0,255,0),cv2.FILLED)
    
plt.figure()
plt.imshow(img)
plt.axis("off")
```
